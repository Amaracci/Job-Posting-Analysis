```python
import pandas as pd
import numpy as np
```

```python
#Loading dataset
data = pd.read_csv(r'Downloads/Uncleaned_DS_jobs.csv')
data
```

```python
data.info()
```

```python
#Dropping the index column because its not necessary
data = data.drop(columns = ['index'])
```

```python
#Removing Glassdoor est. from the salary
data['Salary Estimate'] = data['Salary Estimate'].str.replace(r'\(.*?\)', '', regex=True)
data['Salary Estimate'] = data['Salary Estimate'].str.replace('K', '').str.replace('$','').str.strip()
data['Salary Estimate'].unique()
```

```python
#Sepearating the Minimum Salary from the Maximum Salary
data[['Minimum Salary ($)', 'Maximum Salary ($)']] = data['Salary Estimate'].str.split('-', expand=True)
data['Minimum Salary ($)'] = pd.to_numeric(data['Minimum Salary ($)'], errors='coerce')
data['Maximum Salary ($)'] = pd.to_numeric(data['Maximum Salary ($)'], errors='coerce')
data = data.drop('Salary Estimate', axis = 1)
```

```python
#Moving the Min & Max Salary columns to the 3rd and 4th columns

#Getting the list order
cols = data.columns.tolist()

#Removing the Min & Max Salary columns from the current position
cols.remove('Minimum Salary ($)')
cols.remove('Maximum Salary ($)')

#Putting them in the order i want
cols.insert(1, 'Minimum Salary ($)')
cols.insert(2, 'Maximum Salary ($)')

#Reindexing the dataframe to the new order
data = data[cols]
data
```

```python
#Removing extra and unwanted characters from job description
data['Job Description'] = data['Job Description'].str.encode('ascii', 'ignore').str.decode('ascii')
```

```python
#Removing the ratings attached to the company names
data['Company Name'] = data['Company Name'].str.extract(r'^(.*?)(?:\s\d\.\d)?$')[0]
data
```

```python
#Replacing -1 in the rating with np.nan to help with filling missing rates
data['Rating'] = data['Rating'].replace(-1, np.nan)

# Filling the missing ratings with the average rating
data['Rating'] = data['Rating'].fillna(data['Rating'].mean()).round(1)
data
```

```python
(data['Company Name'] == '-1').sum() #I noticed -1 was used to reperesent unknown values, i did this how many was missen, company name and location result was both 0
(data['Location'] == '-1').sum() 
(data['Headquarters'] == '-1').sum() # This had 31 missen values

#Replacing '-1' with Unknown for analysis
data['Headquarters'] = data['Headquarters'].replace('-1', 'Unknown')
data
```

```python
# CLeaning the Size Column for proper analysis
data['Size'].unique() #This helps me get the unique size categories

#Defining the group labels
size_labels = {
    '1 to 50 employees': 'Micro',
    '51 to 200 employees': 'Small',
    '201 to 500 employees': 'Medium',
    '501 to 1000 employees': 'Medium Large',
    '1001 to 5000 employees': 'Large',
    '5001 to 10000 employees': 'Very Large',
    '10000+ employees': 'Enterprise',
    '-1': 'Unknown',
    'Unknown': 'Unknown'
}
data['Size'] = data['Size'].map(size_labels) #This maps the group levels into the size column
data
```

```python
(data['Founded'] == -1).sum() # We have 118 missen values reperesented as -1

data['Founded'] = data['Founded'].replace(-1, np.nan)
data['Founded'] = pd.to_numeric(data['Founded'], errors = 'coerce')
data['Founded'] = data['Founded'].astype('Int64')
#data['Founded'] = data['Founded'].fillna('Unknown')
data
```

```python
(data['Type of ownership'] == '-1').sum() #This had 27 missen values as -1

data['Type of ownership'].unique() #This was to know the unique types of ownership being recorded

#Defining the types as some where repeated but in another format
ownership_mapping = {
    'Company - Private': 'Private',
    'Private Practice / Firm': 'Private',
    'Company - Public': 'Public',
    'Subsidiary or Business Segment': 'Subsidiary',
    'Nonprofit Organization': 'Nonprofit',
    'Government': 'Government',
    'Hospital': 'Hospital',
    'College / University': 'Education',
    'Self-employed': 'Self-employed',
    'Contract': 'Contract',
    'Other Organization': 'Other',
    'Unknown': 'Unknown',
    '-1': 'Unknown'
}

data['Type of ownership'] = data['Type of ownership'].map(ownership_mapping) #This inserts the maps in the ownership types
data
```

```python
(data['Industry'] == '-1').sum() #This had 71 missen values as -1

data['Industry'].unique()

data['Industry'] = data['Industry'].replace('-1', 'Unknown') #Replacing -1 with Unknown
data
```

```python
(data['Sector'] == '-1').sum() #This had 71 missen values as -1

data['Sector'].unique()

data['Sector'] = data['Sector'].replace('-1', 'Unknown') #Replacing -1 with Unknown
data
```

```python
(data['Revenue'] == '-1').sum() #This had 27 missen values as -1

data['Revenue'].unique()

# Defining Categorical Values for a better understanding of the revenue Values
revenue_categories = {
    'Less than $1 million (USD)': 'Under $1M',
    '$1 to $5 million (USD)': '$1M–$5M',
    '$5 to $10 million (USD)': '$5M–$10M',
    '$10 to $25 million (USD)': '$10M–$25M',
    '$25 to $50 million (USD)': '$25M–$50M',
    '$50 to $100 million (USD)': '$50M–$100M',
    '$100 to $500 million (USD)': '$100M–$500M',
    '$500 million to $1 billion (USD)': '$500M–$1B',
    '$1 to $2 billion (USD)': '$1B–$2B',
    '$2 to $5 billion (USD)': '$2B–$5B',
    '$5 to $10 billion (USD)': '$5B–$10B',
    '$10+ billion (USD)': 'Over $10B',
    'Unknown / Non-Applicable': 'Unknown',
    '-1': 'Unknown'
}
data['Revenue'] = data['Revenue'].map(revenue_categories).str.strip()
data
```

```python
(data['Competitors'] == '-1').sum() #This had 501 missing values as -1

data['Competitors'] = data['Competitors'].replace('-1', 'Unknown').str.strip() #Replacing -1 with Unknown
data
```
